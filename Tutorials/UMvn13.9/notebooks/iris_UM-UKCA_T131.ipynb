```python
import iris
import iris.quickplot as qplt
import numpy as np
import matplotlib.pyplot as plt
iris.FUTURE.save_split_attrs = True
```

```python
# read-in fields including Sec_Org product
# YOU WILL NEED TO CHANGE THESE PATHS FOR YOUR FILES
aod_ctrl=iris.load_cube('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/iris_Task131_AOD_ctrl.nc','atmosphere_optical_thickness_due_to_aerosol')
ssa_ctrl=iris.load_cube('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/iris_Task131_SSA_ctrl.nc','single_scattering_albedo_in_air_due_to_ambient_aerosol_particles')
toa_ctrl=iris.load_cube('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/iris_Task131_TOA_ctrl.nc','toa_net_downward_radiative_flux')
```

```python
# read-in fields without Sec_Org product
# YOU WILL NEED TO CHANGE THESE PATHS FOR YOUR FILES
aod_expt=iris.load_cube('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/iris_Task131_AOD_expt.nc','atmosphere_optical_thickness_due_to_aerosol')
ssa_expt=iris.load_cube('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/iris_Task131_SSA_expt.nc','single_scattering_albedo_in_air_due_to_ambient_aerosol_particles')
toa_expt=iris.load_cube('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/iris_Task131_TOA_expt.nc','toa_net_downward_radiative_flux')
```

```python
# difference the respective fields
aoddiff=aod_expt.copy()
aoddiff.data=np.zeros(aoddiff.data.shape)
aoddiff.data=aod_expt.data-aod_ctrl.data

ssadiff=ssa_expt.copy()
ssadiff.data=np.zeros(ssadiff.data.shape)
ssadiff.data=ssa_expt.data-ssa_ctrl.data

toadiff=toa_expt.copy()
toadiff.data=np.zeros(toadiff.data.shape)
toadiff.data=toa_expt.data-toa_ctrl.data
```

```python
# save the differences for viewing later
iris.save(aoddiff, '/home/vagrant/iris_Task131_AOD_diff.nc', netcdf_format='NETCDF4')
iris.save(ssadiff, '/home/vagrant/iris_Task131_SSA_diff.nc', netcdf_format='NETCDF4')
iris.save(toadiff, '/home/vagrant/iris_Task131_TOA_diff.nc', netcdf_format='NETCDF4')
```

```python
# plot the AOD
qplt.pcolormesh(aoddiff)
plt.gca().coastlines()
```

```python
# plot the SSA
qplt.pcolormesh(ssadiff)
plt.gca().coastlines()
```

```python
# plot the TOA
qplt.pcolormesh(toadiff)
plt.gca().coastlines()
```
