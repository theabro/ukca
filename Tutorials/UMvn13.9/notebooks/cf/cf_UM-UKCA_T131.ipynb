```python
# need to import cfplot before cf due to issues with matplotlib and esmpy
import cfplot as cfp
import cf
import numpy as np
```

```python
# read-in fields including Sec_Org product
aod_ctrl=cf.read('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/cf_Task131_AOD_ctrl.nc',select='atmosphere_optical_thickness_due_to_aerosol')[0]
ssa_ctrl=cf.read('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/cf_Task131_SSA_ctrl.nc',select='single_scattering_albedo_in_air_due_to_ambient_aerosol_particles')[0]
toa_ctrl=cf.read('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/cf_Task131_TOA_ctrl.nc',select='toa_net_downward_radiative_flux')[0]
```

```python
# read-in fields without Sec_Org product
aod_expt=cf.read('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/cf_Task131_AOD_expt.nc',select='atmosphere_optical_thickness_due_to_aerosol')[0]
ssa_expt=cf.read('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/cf_Task131_SSA_expt.nc',select='single_scattering_albedo_in_air_due_to_ambient_aerosol_particles')[0]
toa_expt=cf.read('/home/vagrant/Tutorials/UMvn13.9/sample_output/Task13.1/cf_Task131_TOA_expt.nc',select='toa_net_downward_radiative_flux')[0]
```

```python
# difference the respective fields
aoddiff=aod_expt-aod_ctrl
ssadiff=ssa_expt-ssa_ctrl
toadiff=toa_expt-toa_ctrl
```

```python
# save the differences for viewing later
cf.write(aoddiff, '/home/vagrant/cf_Task131_AOD_diff.nc', fmt='NETCDF4')
cf.write(ssadiff, '/home/vagrant/cf_Task131_SSA_diff.nc', fmt='NETCDF4')
cf.write(toadiff, '/home/vagrant/cf_Task131_TOA_diff.nc', fmt='NETCDF4')
```

```python
# plot the AOD
# error with using blockfill_fast, so use blockfill
cfp.con(aoddiff, blockfill=True, lines=False, title='Difference in Aerosol Optical Depth at 550nm')
```

```python
# plot the SSA
# error with using blockfill_fast, so use blockfill
cfp.con(ssadiff, blockfill=True, lines=False, title='Difference in Single Scattering Albedo at 550nm')
```

```python
# plot the TOA
# error with using blockfill_fast, so use blockfill
cfp.con(toadiff, blockfill=True, lines=False, title='Difference in Top of Atmosphere Net Downward Radiative Flux')
```
